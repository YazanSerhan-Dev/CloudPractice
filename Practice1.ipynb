```python
from google.colab import drive
drive.mount('/content/drive')

import ipywidgets as widgets
from IPython.display import display, clear_output

file_path = '/content/drive/MyDrive/students.txt'


# ---------- Read students from file ----------
def load_students():
    students = []

    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            parts = line.split(',', 4)
            if len(parts) != 5:
                continue

            first_name = parts[0].strip()
            last_name = parts[1].strip()
            email = parts[2].strip()
            courses_str = parts[3].strip()
            link = parts[4].strip()

            courses = {}
            if courses_str:
                for item in courses_str.split(';'):
                    if ':' in item:
                        course, grade = item.split(':', 1)
                        try:
                            courses[course.strip()] = int(grade.strip())
                        except ValueError:
                            pass

            students.append({
                'first_name': first_name,
                'last_name': last_name,
                'email': email,
                'courses': courses,
                'link': link
            })

    return students


students = load_students()


# ---------- Helper ----------
def courses_to_text(courses_dict):
    if not courses_dict:
        return "No courses"
    return "\n".join([f"{course}: {grade}" for course, grade in courses_dict.items()])


def student_to_line(student):
    courses_str = ";".join([f"{course}:{grade}" for course, grade in student['courses'].items()])
    return f"{student['first_name']},{student['last_name']},{student['email']},{courses_str},{student['link']}\n"


# ---------- Widgets ----------
title = widgets.HTML("<h2>Student Management System</h2>")

student_dropdown = widgets.Dropdown(
    options=[(f"{s['first_name']} {s['last_name']} - {s['email']}", i) for i, s in enumerate(students)],
    description='Student:',
    layout=widgets.Layout(width='800px')
)

first_name_box = widgets.Text(
    description='First Name:',
    disabled=True,
    layout=widgets.Layout(width='800px')
)

last_name_box = widgets.Text(
    description='Last Name:',
    disabled=True,
    layout=widgets.Layout(width='800px')
)

email_box = widgets.Text(
    description='Email:',
    disabled=True,
    layout=widgets.Layout(width='800px')
)

link_box = widgets.Text(
    description='Link:',
    disabled=True,
    layout=widgets.Layout(width='800px')
)

courses_area = widgets.Textarea(
    description='Courses:',
    disabled=True,
    layout=widgets.Layout(width='800px', height='150px')
)

update_title = widgets.HTML("<h3>Update Selected Student</h3>")

new_course_box = widgets.Text(
    description='Course:',
    placeholder='Enter course name',
    layout=widgets.Layout(width='400px')
)

new_grade_box = widgets.IntText(
    description='Grade:',
    layout=widgets.Layout(width='250px')
)

save_button = widgets.Button(
    description='Add Course and Save',
    button_style='success'
)

output = widgets.Output()


# ---------- Display selected student ----------
def show_student(index):
    student = students[index]
    first_name_box.value = student['first_name']
    last_name_box.value = student['last_name']
    email_box.value = student['email']
    link_box.value = student['link']
    courses_area.value = courses_to_text(student['courses'])


def on_student_change(change):
    if change['name'] == 'value' and change['new'] is not None:
        show_student(change['new'])


student_dropdown.observe(on_student_change, names='value')


# ---------- Save update ----------
def on_save_clicked(b):
    with output:
        clear_output()

        selected_index = student_dropdown.value
        student = students[selected_index]

        course_name = new_course_box.value.strip()
        grade = new_grade_box.value

        if course_name == "":
            print("Please enter a course name.")
            return

        if grade < 0 or grade > 100:
            print("Please enter a valid grade between 0 and 100.")
            return

        # create updated copy
        updated_student = {
            'first_name': student['first_name'],
            'last_name': student['last_name'],
            'email': student['email'],
            'courses': dict(student['courses']),
            'link': student['link']
        }

        updated_student['courses'][course_name] = grade

        # append new line to file (do not delete old line)
        with open(file_path, 'a', encoding='utf-8') as f:
            f.write(student_to_line(updated_student))

        # update current student in memory so UI refreshes
        students[selected_index] = updated_student
        show_student(selected_index)

        new_course_box.value = ""
        new_grade_box.value = 0

        print("Student updated successfully.")
        print("A new line was appended to students.txt.")


save_button.on_click(on_save_clicked)


# ---------- Initial display ----------
if students:
    show_student(0)

display(title)
display(student_dropdown)
display(first_name_box)
display(last_name_box)
display(email_box)
display(link_box)
display(courses_area)
display(update_title)
display(widgets.HBox([new_course_box, new_grade_box]))
display(save_button)
display(output)
```
